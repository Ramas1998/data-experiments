# Sharechat post virality analysis

We scraped 221 Sharechat posts from 8 tags / 2 content buckets on 25th April 2020 and tracked them daily for 7 days. This notebook visualises how the posts grew in terms of virality metrics i.e. views, likes, shares, comments and reposts.

The tags targeted for this analysis were: 'LATEST Corona Update', 'Corona : Pandemic not in control',
       'Corona virus protection', 'Coronavirus worldwide ', 'Covid-19',
       'Corona : Question-Answer', ' National-International News',
       ' News and newspaper clip'

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/Users/kruttikanadig/Documents/Tattle/sharechat-scraper/virality_df_2020-05-02.csv", index_col=0)
```

```python
data
```

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
data['tag_translation'] = data['tag_translation'].map(lambda x: deEmojify(x))
```

```python
def generate_post_virality_plots(data):
    metrics = ["views", "likes", "external_shares", "comments", "reposts"]
    fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize = (20, 40))
    row = 0
    data['tag_translation'] = data['tag_translation'].map(lambda x: deEmojify(x))
    for metric in metrics:
        data_metric = data[['post_permalink', 'scraped_date', 'bucket_name', 'bucket_id', 'tag_name', 
                            'tag_genre', 'tag_translation', metric, metric+'_t+1', metric+'_t+2', metric+'_t+3', 
                            metric+'_t+4', metric+'_t+5', metric+'_t+6', metric+'_t+7']]
        data_metric.rename(columns= {metric: metric+"_t+0"}, inplace=True)
        data_metric_long = data_metric.reset_index().melt(id_vars = ["post_permalink", "index", 'scraped_date', 
                                                                 'bucket_name', 'bucket_id', 'tag_name', 'tag_genre',
                                                                    'tag_translation'],
                                                         value_vars = [metric+"_t+0", metric+'_t+1', metric+'_t+2', 
                                                                       metric+'_t+3', metric+'_t+4', 
                                                                       metric+'_t+5', metric+'_t+6', metric+'_t+7'],
                                                         var_name = metric).sort_values([
                                                        "post_permalink", metric]).reset_index(drop=True)
        sns.lineplot(data=data_metric_long, x=metric, y="value", estimator = "mean", ax=ax[row, 0])
        ax[row, 0].set_title("Average"+" "+metric, fontsize=16)
        ax[row, 0].tick_params(labelsize=14, rotation = 45)
        ax[row, 0].set_ylabel("Count", fontsize = 14, rotation = 90)
        ax[row, 0].set_xlabel(metric, fontsize = 14)
        sns.catplot(x = metric, y = "value", data=data_metric_long, jitter = "0.25", hue = "tag_translation", ax=ax[row, 1])
        ax[row, 1].set_title(metric+" "+"on individual posts", fontsize=16)
        ax[row, 1].tick_params(labelsize=14, rotation = 45)
        ax[row, 1].set_ylabel("Count", fontsize = 14, rotation = 90)
        ax[row, 1].set_xlabel(metric, fontsize = 14)
        plt.close(2)
        plt.close(3)
        row+=1
    fig.suptitle("Virality metrics for 221 Sharechat posts over 7 days", size = 20)
    fig.tight_layout()
    fig.subplots_adjust(top=0.95)
    plt.savefig("post_virality.png")
    fig.show()
           



```

```python
generate_post_virality_plots(data)
```

```python
def generate_tag_virality_plots(data):
    metrics = ["views", "likes", "external_shares", "comments", "reposts"]
    fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize = (20, 40))
    row = 0
    data['tag_translation'] = data['tag_translation'].map(lambda x: deEmojify(x))
    for metric in metrics:
        data_metric = data[['post_permalink', 'scraped_date', 'bucket_name', 'bucket_id', 'tag_name', 
                            'tag_genre', 'tag_translation', metric, metric+'_t+1', metric+'_t+2', metric+'_t+3', 
                            metric+'_t+4', metric+'_t+5', metric+'_t+6', metric+'_t+7']]
        data_metric.rename(columns= {metric: metric+"_t+0"}, inplace=True)
        data_metric_long = data_metric.reset_index().melt(id_vars = ["post_permalink", "index", 'scraped_date', 
                                                                 'bucket_name', 'bucket_id', 'tag_name', 'tag_genre',
                                                                    'tag_translation'],
                                                         value_vars = [metric+"_t+0", metric+'_t+1', metric+'_t+2', 
                                                                       metric+'_t+3', metric+'_t+4', 
                                                                       metric+'_t+5', metric+'_t+6', metric+'_t+7'],
                                                         var_name = metric).sort_values([
                                                        "post_permalink", metric]).reset_index(drop=True)
        sns.lineplot(data=data_metric_long, x=metric, y="value", hue = "tag_translation", estimator = "mean", ci=None, ax=ax[row, 0])
        ax[row, 0].set_title("Average"+" "+metric+" per tag", fontsize=16)
        ax[row, 0].tick_params(labelsize=14, rotation = 45)
        ax[row, 0].set_ylabel("Count", fontsize = 14, rotation = 90)
        ax[row, 0].set_xlabel(metric, fontsize = 14)
        sns.lineplot(data=data_metric_long, x=metric, y="value", hue = "tag_genre", estimator = "mean", ci=None, ax=ax[row, 1])
        ax[row, 1].set_title("Average"+" "+metric+" per genre", fontsize=16)
        ax[row, 1].tick_params(labelsize=14, rotation = 45)
        ax[row, 1].set_ylabel("Count", fontsize = 14, rotation = 90)
        ax[row, 1].set_xlabel(metric, fontsize = 14)
        plt.close(2)
        plt.close(3)
        row+=1
    fig.suptitle("Virality metrics for 221 Sharechat posts over 7 days", size = 20)
    fig.tight_layout()
    fig.subplots_adjust(top=0.95)
    plt.savefig("tag_virality.png")
    fig.show()
```

```python
generate_tag_virality_plots(data)
```

## Observations

- The biggest growth happens in the first 24 hours 
- Likes and external shares (to Whatsapp etc) are more popular forms of engagement than comments and reposts within Sharechat
- Comments and reposts are negligible 
- There are several exteme outliers (posts with much higher virality than average)
- The outlier with the highest views (40k) belongs to the "National-International News" tag
- Outliers with the highest likes (800+) and external shares (175+) belong to Coronavirus related tags
- The tag "LATEST Corona Updates" got the most views, followed by "National-International News". The former also got more comments than other tags
- The outlier with the most comments (300+) belongs to the "LATEST Corona Updates" tag
- The tag "Corona: Question-Answer" had the most external shares
- In this data sample, Coronavirus content (from 6 tags classified under Health & Fitness) got more engagement than general News content from 2 tags, however, the News content got more views. Patterns in engagement and views show what type of content was consumed more actively vs. more passively.
